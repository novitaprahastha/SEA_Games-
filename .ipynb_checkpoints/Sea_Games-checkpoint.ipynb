```python
import pandas as pd 
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests
import html5lib
```

# SOAL 3

```python
url = requests.get("https://www2.2019seagames.com/countries/")
data = BeautifulSoup(url.content, "html.parser")

data1 = []

for i in data.find_all("em"):
    data1.append(i.text)

data1
```

```python
asean = data1[::6]

asean = asean[0:11]

asean
```

```python
gold= data1[2::6]
# gold

gold1 = []
gold2 = []

for i in range(len(gold)):
    temp = gold[i].split(" ")
    gold1.append(temp[1])
    
for j in range(len(gold1)):
    temp = int(gold1[j])
    gold2.append(temp)
    
gold2 = gold2[:11]

gold2
```

```python
url1 = requests.get("https://rs.2019seagames.com/RS2019/mobiapp/MedalTally")
data = BeautifulSoup(url1.content, "html.parser")
# data

out1 = []

for i in data1.find_all("small"):
    out1.append(i.text)

out1
```

```python
out2 = out1[5:]
out2
```

```python
asean2 = out2[::5]
asean2
```

```python
gold3 = out2[1::5]
gold3
```

```python
asean17 = list(zip(asean, gold2))
asean17.sort()

asean19 = list(zip(asean2, gold3))
asean19.sort()
```

```python
# asean17
# asean19
```

```python
asean_all = []

for i in range(len(asean19)):
    temp = asean19[i]
    asean_all.append(temp[0])

asean_all
```

```python
gold17 = []

for i in range(len(asean17)):
    temp = asean17[i]
    gold17.append(temp[1])

gold17
```

```python
gold19 = []

for i in range(len(asean19)):
    temp = asean19[i]
    gold19.append(int(temp[1]))

gold19
```

```python
plt.figure(figsize=(15,6))

plt.plot(asean_all, gold17, "r-o")
plt.plot(asean_all, gold19, "g-o")
plt.plot(asean_all[gold17.index(max(gold17))], max(gold17), '*', markersize = 25, markerfacecolor = "y")
plt.plot(asean_all[gold19.index(max(gold19))], max(gold19), '*', markersize = 25, markerfacecolor = "y")
plt.suptitle('GOLD MEDAL TALLY SEA GAMES 2017 & 2019', fontsize = 25)
plt.xticks(rotation = 30)
plt.legend(['Gold 2017', 'Gold 2019', 'Champion 2017', 'Champion 2019'])
plt.grid(True)
plt.show
```

```python
plt.figure(figsize = (15,7))
plt.subplot(1, 2, 1)
plt.pie(gold17, labels = asean_all, autopct='%1.1f%%',startangle=90,counterclock=True,)
plt.title('%GOLD MEDAL TALLY SEA GAMES 2017', fontsize = 20)

plt.subplot(1, 2, 2)
plt.pie(gold19, labels = asean_all, autopct='%1.1f%%',startangle=90,counterclock=True)
plt.title('%GOLD MEDAL TALLY SEA GAMES 2019', fontsize = 20)
plt.show()
```
